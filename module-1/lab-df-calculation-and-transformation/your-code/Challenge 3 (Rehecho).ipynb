```python
import pandas as pd
```

```python
file = pd.read_csv("Orders.csv")
```

```python
data = pd.DataFrame(file)
data.head(50)
```

```python
gasto_cliente = data.groupby('CustomerID').sum().amount_spent
```

```python
clientes = pd.DataFrame(gasto_cliente)
```

```python
clientes.head()
```

```python
def Labels(y):
    if y >= gasto_cliente.quantile(0.95):
        return "VIP" 
    elif y >= gasto_cliente.quantile(0.75):
        return "Preferred"
    else:
        return "Others"

clientes['Labels']= clientes['amount_spent'].apply(Labels)
    
```

```python
clientes.head()
```

```python
new_clients = clientes.drop_duplicates()
```
